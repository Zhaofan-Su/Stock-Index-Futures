```python
file_name = 'IF9999_validation'
data = pd.read_csv(f'./data/{file_name}.csv', index_col=0)
```

```python
data.index = data.index.map(lambda x:str2datetime(x.replace('"','')))
data['close'] = data['close'].map(lambda x:float(x.replace('"','').replace(',','')))
data['open'] = data['open'].map(lambda x:float(x.replace('"','').replace(',','')))
data['high'] = data['high'].map(lambda x:float(x.replace('"','').replace(',','')))
data['low'] = data['low'].map(lambda x:float(x.replace('"','').replace(',','')))
data['volume'] = data['volume'].map(lambda x:float(x.replace('"','').replace('K','')) * 1000)
```
